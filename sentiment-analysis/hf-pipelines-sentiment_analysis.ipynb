# Import deps

```python
from transformers import pipeline
```

# Init pipeline

```python
# distilbert-base-uncased-finetuned-sst-2-english
classifier = pipeline('sentiment-analysis')
```

```python
# another model
classifier = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')
```

# Logic

```python
# example01
output = classifier('I am very happy.')
print(output[0]['label'])

# example02
output = classifier('I really don\'t like what you did')
print(output[0]['label'])

# example03
output = classifier('I really like what you did')
print(output[0]['label'])

# example04
output = classifier('This is good')
print(output[0]['label'])
```

# Testing

```python
input = ""
output = classifier(input)
print(output)
```
