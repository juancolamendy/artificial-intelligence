#Import deps

```python
import fasttext
```

# Train a classification model

```python
model = fasttext.train_supervised('cooking.stackexchange.txt')
```

```python
model.predict("how to cook some rice?")
```

# Save and load the model

```python
model.save_model("models/model_coooking.bin")
```

```python
model = fasttext.load_model("models/model_coooking.bin")
print(model.labels)
```

```python
model.predict("Which baking dish is best to bake a banana bread ?")
```
