# Import dep

```python
from transformers import pipeline
```

# Init pipeline

```python
# Initialize MLM pipeline
mlm = pipeline('fill-mask')

# Get mask token
mask = mlm.tokenizer.mask_token
```

# Define helper functions

```python
def print_result(result):
  print('--- result:')
  for res in result:
    print(res)
```

# Logic

```python
# Get result for particular masked phrase
phrase = f'Read the rest of this {mask} to understand things in more detail'
result = mlm(phrase)

# Print result
print_result(result)
```

```python
# Get result for particular masked phrase
phrase = f'you will never know your niche without knowing {mask}'
result = mlm(phrase)

# Print result
print_result(result)
```
